```python
import glob
import re
import itertools
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = glob.glob("./*.txt")[0]
```

```python
ls = []
with open(file, 'r') as f:
    lines = f.readlines()
    for j in range(len(lines)):
        if lines[j] != "\n":
            ls.append(lines[j])
```

```python
def grouper(n, iterable):
    args = [iter(iterable)]*n
    return ([e for e in t if e!=None ] for t in itertools.zip_longest(*args))
```

```python
gen=grouper(101, ls)
```

```python
ls2 = []
for i in gen:
    ls2.append(i)
```

```python
ls2[0]
```

```python
ls3 = []
for i in range(len(ls2)):
    ls4 = []
    for j in ls2[i]:
        k = re.findall("[0].[0-9][0-9][0-9][0-9][0-9][0-9]",j)
        if k != []:
            for l in k:
                ls4.append(l)
    ls3.append(ls4)
    
```
